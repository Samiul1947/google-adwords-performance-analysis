```python
import csv
import pandas as pd

def key_id(value):
    value = str(value).lower()
    if 'scrum' in value or 'csm' in value or 'smum' in value or 'srum' in value:
        return 1
    elif 'amazon' in value or 'aws' in value or 'devops' in value:
        return 2
    elif 'pmp' in value or 'project management' in value or 'pmi' in value or 'proyectos' in value or 'it project' in value:
        return 3
    elif 'cloud' in value:
        return 4
    elif 'capm' in value:
        return 5
    elif 'cspo' in value:
        return 6
    elif 'itil' in value or 'itl' in value:
        return 7
    elif 'simpl' in value or 'simli' in value or 'simpi' in value or 'smpli' in value or 'simi' in value or 'sipli' in value:
        return 8
    elif 'safe' in value or 'scale' in value:
        return 9
    elif 'togaf' in value or 'udacity' in value or 'it architect' in value:
        return 10
    else:
        return None

# Apply function to DataFrame (NO INDENTATION)
df['Keyword_ID'] = df['Keyword'].apply(key_id)
df.head(10)
```

```python
# Apply the key_id function to assign each row a Keyword ID
df['Keyword_ID'] = df['Keyword'].apply(key_id)
```

```python
# Drop the 'Trends' column if present
if 'Trends' in df.columns:
    df = df.drop('Trends', axis=1)

# Reorder columns (MATCH EXACT COLUMN NAMES)
cols = [
    'Title', 'Keyword', 'Keyword_ID', 'Position', 'Previous position',
    'Last Seen', 'Search Volume', 'CPC', 'Traffic', 'Traffic (%)',
    'Traffic Cost', 'Traffic Cost (%)', 'Competition', 'Number of Results',
    'Keyword Difficulty'
]

df = df[cols]

df.head(10)
```

```python
# Define the mapping keywords
keyword_map = {
    "Scrum Master": ['scrum', 'csm', 'smum', 'srum', 'scrummaster'],
    "AWS": ['amazon', 'aws','devops'],
    "Cloud": ['cloud'],
    "PMP": ['pmp', 'project management', 'pmi', 'proyectos', 'it project'],
    "CAPM": ['capm'],
    "CSPO": ['cspo'],
    "ITIL": ['itil', 'itl'],
    "Simplilearn": ['simplilearn', 'simpli', 'simpi', 'smpli', 'simi', 'sipli', 'simply','simlilearn','simplelern',
                    'simplelearn', 'simplylearn', 'simplileanr','simpllearn','simple','simplearn',],
    "SAFe": ['safe', 'scaled agile', 'scale agile'],
    "TOGAF": ['togaf', 'udacity', 'it architect'],
}

# Function to categorize each row
def classify_keyword(value):
    val = str(value).lower()
    for label, keywords in keyword_map.items():
        if any(k in val for k in keywords):
            return label
    return None  # Or "Other" if you prefer

# Apply to your dataframe
df['Keyword'] = df['Keyword'].apply(classify_keyword)

# Show sample
df.head(10)
```

```python
# Create a lookup table of unique Keyword IDs
keyword_table = (
    df[['Keyword_ID', 'Keyword']]
    .drop_duplicates()
    .sort_values(by='Keyword_ID')
)

keyword_table.to_csv('keyword.csv', index=False)
```

```python
# Aggregate total search volume per keyword_ID
search_volume_df = (
    df.groupby('Keyword_ID', as_index=False)['Search Volume']
      .sum()
      .rename(columns={
          'Keyword_ID': 'keyword_ID',
          'Search Volume': 'search_volume'
      })
)
```

```python
search_volume_df.to_csv(
    r"C:\Users\ADMIN\Downloads\search_volume.csv",
    index=False
)
```

```python
# Calculate average keyword difficulty per keyword_ID
keyword_difficulty_df = (
    df.groupby('Keyword_ID', as_index=False)['Keyword Difficulty']
      .mean()
      .rename(columns={
          'Keyword_ID': 'keyword_ID',
          'Keyword Difficulty': 'avg_difficulty'
      })
)

# Assign difficulty level using IF logic
keyword_difficulty_df['difficulty_level'] = keyword_difficulty_df['avg_difficulty'].apply(
    lambda x: 'Hard' if x >= 50 else 'Moderate'
)
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.replace('%', 'percent')
    .str.replace(' ', '_')
    .str.lower()
)
```

```python
# Create traffic_percent if missing
if 'traffic_percent' not in df.columns:
    df['traffic_percent'] = (df['traffic'] / df['traffic'].sum()) * 100

# Create traffic_cost_percent if missing
if 'traffic_cost_percent' not in df.columns:
    df['traffic_cost_percent'] = (
        df['traffic_cost'] / df['traffic_cost'].sum()
    ) * 100
```

```python
if 'keyword_id' not in df.columns:
    df['keyword_id'] = df['keyword'].astype('category').cat.codes + 1
```
